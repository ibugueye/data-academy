### **1. Architecture Générale de l'Application**

Pour que la formation soit efficace, elle doit suivre une pédagogie progressive :
*   **Niveau 1 : Comprendre (Théorie)** – Explication des concepts.
*   **Niveau 2 : Visualiser (Démonstration)** – Voir le résultat en action.
*   **Niveau 3 : Pratiquer (Labo de Code)** – Modifier les paramètres et voir l'impact.
*   **Niveau 4 : Simuler (Business Case)** – Prendre des décisions virtuelles.

### **2. Décomposition des Modules de Formation**

Voici comment structurer les 5 points de votre programme au sein de l'application.

#### **Module 1 : Les Prévisions Budgétaires Intelligentes**
*   **Objectif Pédagogique :** Comprendre l'écart entre une prévision statique (Excel) et une prévision dynamique (ML).
*   **Scénario interactif :**
    *   L'apprenant dispose d'un historique de ventes fictives.
    *   *Exercice :* Il ajuste des curseurs (ex: "Budget Marketing", "Météo moyenne") et visualise en temps réel la courbe de prévision changer grâce au modèle de régression intégré.
*   **Quiz :** "Si la météo passe de 20°C à 25°C, de combien la prévision augmente-t-elle ?"

#### **Module 2 : Machine Learning au service de la Rentabilité**
*   **Objectif Pédagogique :** Apprendre à classifier les risques automatiquement.
*   **Scénario interactif :**
    *   Présentation d'une liste de projets en cours.
    *   *Exercice :* L'apprenant lance l'algorithme de classification. L'application colorie les projets en "Vert" (rentable) ou "Rouge" (risqué).
    *   *Focus Interprétation :* L'app affiche une "Feature Importance" (ex: 70% du risque vient de la durée du projet).
*   **Lien Contrôle de Gestion :** "Quelle action corrective recommanderiez-vous pour le Projet B identifié comme risqué ?"

#### **Module 3 : Optimisation des Processus**
*   **Objectif Pédagogique :** Résoudre des problèmes de contraintes (minimiser les coûts sous contraintes).
*   **Scénario interactif :**
    *   Un mini-jeu de "Logistique". L'apprenant doit livrer des marchandises.
    *   *Défis :* Il essaie manuellement de trouver le meilleur itinéraire (coût estimé).
    *   *Révélation :* Il clique sur "Lancer l'Optimisation" et voit l'algorithme trouver un trajet 15% moins cher que sa solution manuelle.
*   **Concept clé :** Illustrer la notion de "Gaspillage" réduit par la data.

#### **Module 4 : Data Visualization & Storytelling**
*   **Objectif Pédagogique :** Passer des données brutes au tableau de bord décisionnel.
*   **Scénario interactif :**
    *   L'écran est divisé en deux : à gauche une base de données SQL/Excel brute, à droite une grille vide.
    *   *Exercice Drag & Drop :* L'apprenant doit glisser les champs "Ventes" et "Région" pour créer un graphique.
    *   L'application simule l'interface de Power BI ou Tableau.
*   **Résultat :** Si le KPI "Marge" tombe sous un seuil, une alerte visuelle (couleur rouge) apparaît automatiquement.

#### **Module 5 : Stratégie & Prise de Décision**
*   **Objectif Pédagogique :** Synthétiser les modules précédents pour prendre une décision finale.
*   **Scénario interactif (Simulation de rôle) :**
    *   L'apprenant est "Directeur Financier".
    *   Situation : 3 marchés potentiels (A, B, C). Chaque marché a des données brutes (accessibles via les modules précédents).
    *   *Tâche :* Il doit analyser les données (ROI prédictif, risques) et "valider" un investissement.
    *   *Feedback :* L'application simule le futur (1 an plus tard) et révèle si la décision a été rentable ou non, avec une explication "Pourquoi cela a réussi/échoué".

### **4. Exemple de Fonctionnalité "Live" (Code Streamlit)**

Voici un exemple de ce à quoi pourrait ressembler le code pour le **Module 1** dans votre application Streamlit. Cela permettrait à l'étudiant de faire varier la promotion et de voir la prévision changer instantanément.

```python
import streamlit as st
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Configuration de la page
st.title("Module 1 : Simulateur de Prévisions")
st.write("Ajustez les variables pour voir l'impact sur la prévision des ventes.")

# 1. Génération de données fictives pour l'exercice
@st.cache_data
def load_data():
    np.random.seed(42)
    data = pd.DataFrame({
        'Promotions': np.random.randint(1, 10, 100),
        'Météo': np.random.randint(10, 30, 100),
        'Ventes': np.random.randint(100, 500, 100) + (np.random.randint(1, 10, 100) * 20)
    })
    return data

data = load_data()

# 2. Entraînement du modèle (caché à l'utilisateur pour la simplicité, ou visible en mode "Code")
X = data[['Promotions', 'Météo']]
y = data['Ventes']
model = LinearRegression()
model.fit(X, y)

# 3. Interface Interactive (Sliders)
st.sidebar.header("Paramètres d'Entrée")
promo_input = st.sidebar.slider("Niveau de Promotion (1-10)", 1, 10, 5)
meteo_input = st.sidebar.slider("Température Météo (°C)", 10, 30, 20)

# 4. Prédiction
prediction = model.predict([[promo_input, meteo_input]])[0]

# 5. Affichage du Résultat
st.subheader("Résultat de la Prévision")
st.metric(label="Ventes Prévues (Unités)", value=f"{int(prediction)}")

# Visualisation des données historiques vs prédiction
st.scatter_chart(data, x='Promotions', y='Ventes', color='Météo', title="Données Historiques")
st.info(f"Le modèle prédit {int(prediction)} ventes avec ces paramètres.")
```

### **Module 1 : Prévisions Budgétaires et Saisonnalité**

#### **Contexte Tropicalisé**
Les marchés tropicaux sont fortement influencés par les saisons (saison sèche vs saison des pluies) et les fêtes spécifiques (Ramadan, Tabaski, Fêtes de fin d'année).
*   **Entreprise :** "Naija Beverages Ltd" (Fabricant de jus de fruits et sodas).
*   **Problème :** Les ventes chutent en saison des pluies mais explosent pendant les fêtes religieuses.

#### **Adaptation du Code**
Nous remplaçons la variable "Météo" par "Saison" et ajoutons une variable binaire pour "Période de Fête".

```python
import streamlit as st
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

def app():
    st.title("Module 1 : Prévisions & Impact Saisonnier 🌧️☀️")
    st.write("### Contexte : Naija Beverages Ltd (Afrique de l'Ouest)")

    # 1. Données simulées tropicalisées
    np.random.seed(42)
    data = pd.DataFrame({
        'Budget_Marketing_KCFA': np.random.randint(500, 5000, 24), # Budget en milliers de FCFA
        'Saison': np.random.choice(['Sèche', 'Pluie'], 24), 
        'Fête': np.random.choice([0, 1], 24), # 1 = Ramadan/Tabaski/Noel
        'Ventes': np.random.randint(10000, 50000, 24)
    })
    
    # Encodage manuel simple pour la démo
    data['Est_Saison_Seche'] = (data['Saison'] == 'Sèche').astype(int)
    data['Ventes'] = data['Ventes'] + (data['Est_Saison_Seche'] * 5000) + (data['Fête'] * 15000)

    # Modèle
    X = data[['Budget_Marketing_KCFA', 'Est_Saison_Seche', 'Fête']]
    y = data['Ventes']
    model = LinearRegression().fit(X, y)

    # Interface
    st.sidebar.header("Paramètres Locaux")
    budget = st.sidebar.slider("Budget Marketing (K FCFA)", 500, 10000, 2000)
    saison = st.sidebar.radio("Saison à venir", ['Sèche', 'Pluie'])
    fete = st.sidebar.checkbox("Période de Fête (Ramadan/Tabaski/Noël) ?")

    # Calcul
    est_seche = 1 if saison == 'Sèche' else 0
    prediction = model.predict([[budget, est_seche, int(fete)]])[0]

    # Affichage en FCFA
    st.metric(label="Prévision des Ventes", value=f"{int(prediction):,} FCFA")
    
    st.info("""
    💡 **Insight Local :** Notez l'impact massif des fêtes religieuses sur la consommation 
    de boissons, indépendamment du budget marketing.
    """)

if __name__ == "__main__":
    app()
```

### **Module 2 : Analyse des Coûts et Projet Agricole**

#### **Contexte Tropicalisé**
L'agriculture et l'agro-industrie sont des piliers économiques. Les coûts varient selon la disponibilité des matières premières locales et la logistique.
*   **Entreprise :** "TropiCashew" (Transformation de Noix de Cajou).
*   **Problème :** Identifier les projets d'exportation rentables en tenant compte de la variabilité du prix de la matière première et de la qualité des routes.

#### **Adaptation du Code**
Prédiction de la rentabilité basée sur le prix d'achat local aux producteurs et l'état de la route vers le port.

```python
import streamlit as st
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

def app():
    st.title("Module 2 : Rentabilité de la Filière Agricole 🥜")
    
    # Données
    data = pd.DataFrame({
        'Prix_Achat_Paysan': [350, 400, 380, 450, 300, 420], # FCFA/kg
        'Qualite_Route': ['Bon', 'Mauvais', 'Moyen', 'Mauvais', 'Bon', 'Moyen'],
        'Rendement_Production': [0.4, 0.3, 0.35, 0.25, 0.45, 0.38], # Taux de restitution
        'Rentable': [1, 0, 1, 0, 1, 1]
    })
    
    # Encodage
    map_route = {'Bon': 2, 'Moyen': 1, 'Mauvais': 0}
    data['Qualite_Route_Num'] = data['Qualite_Route'].map(map_route)
    
    X = data[['Prix_Achat_Paysan', 'Qualite_Route_Num', 'Rendement_Production']]
    y = data['Rentable']
    clf = DecisionTreeClassifier().fit(X, y)

    st.write("### Simulation d'un nouveau projet d'achat :")
    prix = st.slider("Prix d'achat aux producteurs (FCFA/kg)", 200, 600, 350)
    route = st.select_slider("État de la route vers l'usine", options=['Mauvais', 'Moyen', 'Bon'])
    rendement = st.slider("Rendement prévu (%)", 20, 50, 35) / 100

    # Prédiction
    route_num = map_route[route]
    prediction = clf.predict([[prix, route_num, rendement]])[0]

    if prediction == 1:
        st.success("✅ Projet Rentable. Les coûts de transport sont maîtrisés.")
    else:
        st.error("🚨 Projet à Risque. La dégradation de la route augmente trop le coût de revient.")

    st.warning("⚠️ **Note :** Une route mauvaise augmente le temps de trajet, la consommation de carburant et l'usure du camion.")

if __name__ == "__main__":
    app()
```

### **Module 3 : Optimisation de la Logistique Urbaine (Moto-Taxis)**

#### **Contexte Tropicalisé**
Dans les grandes villes tropicales (ex: Lagos, Abidjan, Kinshasa), les embouteillages sont monstres et la livraison se fait souvent par moto (deux-roues).
*   **Entreprise :** "JijiExpress" (Service de livraison e-commerce).
*   **Problème :** Optimiser le nombre de livreurs (motos) par quartier pour minimiser les retards.

#### **Adaptation du Code**
Optimisation linéaire pour répartir une flotte de motos entre des quartiers "Centres-Villes" (congestionnés) et "Banlieues" (routes longues mais fluides).

```python
import streamlit as st
from pulp import *

def app():
    st.title("Module 3 : Gestion de Flotte 'Moto-Taxi' 🛵")
    
    st.write("### Objectif : Minimiser les retards de livraison")

    # Données
    quartiers = ["Centre-Ville", "Banlieue_Nord", "Zone_Industrielle"]
    motos_disponibles = 50
    
    # Coût du retard par moto (arbitraire : Centre-Ville embouteillé = beaucoup de retard)
    # On veut MINIMISER le temps total de trajet
    temps_trajet = {
        "Centre-Ville": 30, # min par livraison
        "Banlieue_Nord": 45, 
        "Zone_Industrielle": 20
    }
    
    # Demande de livraisons
    demande = {"Centre-Ville": 15, "Banlieue_Nord": 10, "Zone_Industrielle": 25}

    # Résolution
    prob = LpProblem("Gestion_Flotte_Moto", LpMinimize)
    motos_vars = LpVariable.dicts("Motos", quartiers, 0, None, LpInteger)

    # Fonction objectif : Minimiser le temps total (Temps * Nombre de motos)
    prob += lpSum([motos_vars[q] * temps_trajet[q] for q in quartiers])

    # Contraintes
    prob += lpSum([motos_vars[q] for q in quartiers]) <= motos_disponibles, "TotalMotos"
    for q in quartiers:
        prob += motos_vars[q] >= demande[q] * 0.8, f"Demande_{q}" # Satisfaire au moins 80%

    prob.solve()

    st.write("#### Allocation optimale des motos :")
    for q in quartiers:
        st.write(f"- **{q}** : {int(motos_vars[q].value())} motos")
    
    st.metric("Temps de trajet total optimisé", f"{int(value(prob.objective))} minutes")

if __name__ == "__main__":
    app()
```

### **Module 4 : Tableau de Bord Commercial (Grands Magasins)**

#### **Contexte Tropicalisé**
Suivi des ventes de produits de première nécessité (Riz, Huile, Sucre) ou de produits technologiques (Smartphones) dans un réseau de boutiques.
*   **Entreprise :** "MarchéPlus".
*   **Problème :** Visualiser les performances par région (ex: Nord vs Sud) et par type de paiement (Mobile Money vs Espèces).

#### **Adaptation du Code**
Focus sur les méthodes de paiement locales (Orange Money, Wave, MTN MoMo).

```python
import streamlit as st
import pandas as pd
import plotly.express as px

def app():
    st.title("Module 4 : Suivi Commercial & Mobile Money 📱")
    
    # Données simulées
    data = pd.DataFrame({
        'Region': ['Dakar', 'Abidjan', 'Cotonou', 'Dakar', 'Abidjan'] * 5,
        'Produit': ['Riz', 'Smartphone', 'Huile', 'Smartphone', 'Riz'] * 5,
        'Paiement': ['Mobile Money', 'Cash', 'Mobile Money', 'Cash', 'Cash'] * 5,
        'CA_XOF': [150000, 500000, 80000, 450000, 120000] * 5
    })

    # Filtre Région
    region = st.multiselect("Filtrer par Région", data['Region'].unique(), default=data['Region'].unique())
    df = data[data['Region'].isin(region)]

    # KPIs
    col1, col2 = st.columns(2)
    col1.metric("Chiffre d'Affaires Total", f"{df['CA_XOF'].sum():,} FCFA")
    
    mobile_money_share = (df[df['Paiement'] == 'Mobile Money']['CA_XOF'].sum() / df['CA_XOF'].sum()) * 100
    col2.metric("Part de paiement Mobile", f"{mobile_money_share:.1f}%")

    # Graphique
    fig = px.bar(df, x='Produit', y='CA_XOF', color='Paiement', 
                 title="Ventes par Produit et Mode de Paiement",
                 labels={'CA_XOF': 'Chiffre d\'Affaires (FCFA)'})
    st.plotly_chart(fig)

if __name__ == "__main__":
    app()
```

### **Module 5 : Décision Stratégique (Expansion Minière ou Infrastructure)**

#### **Contexte Tropicalisé**
L'exploitation minière ou les grands travaux sont des secteurs clés. La décision d'investissement dépend de la stabilité politique, du cours des matières premières (Cacao, Or, Pétrole) et du taux de change USD/Local.

#### **Scénario Interactif**
*   **Rôle :** Directeur Investissement.
*   **Mission :** Investir dans un nouveau site minier ou une plantation.
*   **Risque :** Variation du taux de change (Franc local vs Dollar) et instabilité régionale.

#### **Code Simplifié**

```python
import streamlit as st
import numpy as np
import plotly.graph_objects as go

def app():
    st.title("Module 5 : Investissement & Risque de Change 💱")
    
    st.write("#### Contexte : Investissement dans un nouveau gisement d'Or")
    
    # Paramètres
    investissement_usd = st.number_input("Investissement Initial (USD)", value=1_000_000)
    taux_change_local = st.slider("Taux de change actuel (Local/USD)", 500, 1000, 650)
    volatilite_change = st.slider("Volatilité prévue du taux (%)", 5, 30, 10)

    if st.button("Simuler la rentabilité (10 ans)"):
        # Simulation Monte Carlo simple
        sim_results = []
        for _ in range(1000):
            # Simuler l'évolution du taux de change sur 10 ans
            taux_futur = taux_change_local * np.random.normal(1, volatilite_change/100, 10)
            # Revenus aléatoires en USD
            revenus_usd = np.random.normal(200000, 50000, 10).cumsum()
            
            # Conversion en monnaie locale
            revenu_local = revenus_usd[-1] * taux_futur[-1]
            invest_local = investissement_usd * taux_change_local
            
            roi = (revenu_local - invest_local) / invest_local
            sim_results.append(roi)

        # Affichage
        fig = go.Figure()
        fig.add_trace(go.Histogram(x=sim_results, nbinsx=50, name="ROI"))
        st.plotly_chart(fig)
        
        prob_perte = (np.array(sim_results) < 0).mean() * 100
        if prob_perte > 30:
            st.error(f"🚨 Risque élevé ! Probabilité de perte : {prob_perte:.1f}%")
            st.write("La volatilité de la monnaie locale rend le projet hasardeux.")
        else:
            st.success(f"✅ Projet viable. Probabilité de perte : {prob_perte:.1f}%")

if __name__ == "__main__":
    app()
```
